```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (Make sure the image is in the same directory or provide full path)
image = cv2.imread('/content/IMG_20221102_231612.jpg', cv2.IMREAD_GRAYSCALE)

# Ensure the image was loaded successfully
if image is None:
    print("Error: Image not found")
    exit()

# Kernel for morphological operations
kernel = np.ones((5, 5), np.uint8)  # A 5x5 square kernel

# Perform Erosion
erosion = cv2.erode(image, kernel, iterations=1)

# Perform Dilation
dilation = cv2.dilate(image, kernel, iterations=1)

# Display the results
plt.figure(figsize=(10, 7))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Eroded image
plt.subplot(1, 3, 2)
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')
plt.axis('off')

# Dilated image
plt.subplot(1, 3, 3)
plt.imshow(dilation, cmap='gray')
plt.title('Dilation')
plt.axis('off')

plt.tight_layout()
plt.show()
```
